```python
# Paths to 
AMAZON_PATH = "./customers_amazone.json"
PRODUCTS_PATH = "./products.json"
PARTNERS_PATH = "./partners.json"
SUPPLIERS_PATH = "./suppliers.json"
```

```python
# 1 - Install pymongo
# pip install pymongo 
```

```python
# 2 Import pymongo in project
import pymongo
from pymongo import MongoClient
```

```python
# 3 Connect to database
client = MongoClient('localhost', 27017)
```

```python
# 4 - Dbs list
client.list_database_names()
```

```python
# 5 - Use amazon db
db_amazon = client['amazon_db']
```

```python
# 6 - Amazon Customers Collection
amz_coll = db_amazon['customers']
```

```python
import json

# 6 - Inserting customers data
with open(AMAZON_PATH) as file:
    data =  json.load(file)
    amz_coll.insert_many(data)
```

```python
# 7 - Show all documents
for doc in amz_coll.find():
    print(doc)
```

```python
# 8 -Products collection
products_coll = db_amazon['products']

with open(PRODUCTS_PATH) as file:
    data =  json.load(file)
    products_coll.insert_many(data)
```

```python
# 9 - Find Led Zepellin IV
products_coll.find_one({'name':'Led Zepellin IV'}, {'name':1, '_id':1})
```

```python
# 9 - Find Leonard Reys Book
for book in products_coll.find({'book.author_name':'Leonard Reyes'}):
    print(book)
```

```python
# 10 - Inserting partners data
partner_coll = db_amazon['partners']

with open(PARTNERS_PATH) as file:
    data =  json.load(file)
    partner_coll.insert_many(data)
```
